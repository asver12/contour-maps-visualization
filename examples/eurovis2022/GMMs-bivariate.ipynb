```python
# test  test
dist_params = {
    "scatter_points": True,
    "scatter_points_num": int(1000**0.5),
}

vis_params = {
    "marker": "o",
    "edge_colors": False,
    "face_colors": True,
    "alpha": 0.5,  # opacity
    "s":  6**2, # size
}

lim_params = {
    "xlim": [-8.5,16.5],
    "ylim": [-12.5,12.5],
}

colors=color_schemes.get_representiv_colors(color_schemes.get_colorbrewer_schemes())
colorSchemes = color_schemes.get_colorbrewer_schemes()
```

```python
size = 1000 # grid size for raster-dependent visualizations
```

```python
# set common parameters
vis_params['s']=4.5**2
vis_params['alpha']=0.5
vis_params["edge_colors"]=True
#vis_params["face_colors"]=False
dist_params["scatter_points_num"]=int(1500**0.5)
```

# all teaser figures

```python
dist_params["scatter_points_num"]=int(1500**0.5)

pie_params = {
    "pie_charts": True,
    "pie_num": 20,
    "pie_angle": 90,
    "pie_chart_iso_level": 40,
    "pie_chart_contour_method": "equal_density",
    "pie_chart_modus": "light",
}
contour_params = {
    "contours": True,
    "contour_lines": True
}

cross_params = {
    "crosses": True,
    "contour_lines": True
}

scatter_params = dist_params.copy()
scatter_params.update(vis_params)

parameterSets = list(zip(['contour', 'cross', 'sampling', 'pie'],
                         [contour_params, cross_params, scatter_params, pie_params]))
```

```python
# choose data
gaussians_teaser = example_data.generate_four_random_gaussians(size=size)[24][::-1] # Teaser Gaussians
```

```python
figures, axes = [], []
for name, params in parameterSets:
     print(f'generating {name} teaser ... ')
     fig, ax = picture_plot.plot_images([gaussians_teaser], **params, **lim_params)
     print(f'saving {name} teaser ... ')
     fig.savefig(os.path.join(save_path, f'abstract_example_{name}.png'), format='png')
     fig.savefig(os.path.join(save_path, f'abstract_example_{name}.pdf'), format='pdf')
     print(f'... done')
     figures.append(fig)
     axes.append(ax)
```

# scatter plot of mixture and its single components

```python
# choose data
#gaussians_example = example_data.generate_four_random_gaussians(size=size) # Teaser Gaussians
#new_gau = example_data.generate_four_gaussians(size=size)[24]

# manually construct Gaussians
# set all weights to 1. weight parameter not implemented
gaussians = [
    Gaussian.Gaussian(means=[1, 1.5], cov_matrix=[[3.5, 0.5], [0.5, 2]], weight=1.0,
         x_min=-10, x_max=10, y_min=-10, y_max=10, size=size),
    Gaussian.Gaussian(means=[5, -0], cov_matrix=[[3, -3], [-2, 4]], weight=0.5,
         x_min=-10, x_max=10, y_min=-10, y_max=10, size=size),
    Gaussian.Gaussian(means=[1, -2], cov_matrix=[[7, 2], [3, 4]], weight=0.3,
          x_min=-10, x_max=10, y_min=-10, y_max=10, size=size),
    Gaussian.Gaussian(means=[3, 5], cov_matrix=[[5, 0], [0, 5]], weight=1,
         x_min=-10, x_max=10, y_min=-10, y_max=10, size=size),
    Gaussian.Gaussian(means=[-2, 1], cov_matrix=[[2, 4], [1, 1]], weight=0.1,
         x_min=-10, x_max=10, y_min=-10, y_max=10, size=size),
]
```

```python
dist_params["scatter_points_num"]=int(1000**0.5)
vis_params['xlim']=[-8,12]
vis_params['ylim']=[-8.5 , 11.5]

# plot mixture
fig, ax = picture_plot.plot_images([gaussians], **vis_params, **dist_params)
fig.savefig(os.path.join(save_path, f'scatter-all-components'))
#fig.savefig(os.path.join(save_path, f'scatter-all-components.png'), format="png")
```

```python
# plot single components
for idx,gaussian in enumerate(gaussians):
    fig, ax = picture_plot.plot_images([[gaussian]], colors=[colors[idx]], schatter_points_colors=[colorSchemes[idx]],
                                       legend=False, **vis_params, **dist_params)
    fig.savefig(os.path.join(save_path, f'scatter-component-{idx}'))
    fig.savefig(os.path.join(save_path, f'scatter-component-{idx}.png'), format="png")
```
